```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# specify the url to scrape
url_2020 = 'https://web.archive.org/web/20200630235307/https://www.house.gov/representatives'
url_2021 = 'https://web.archive.org/web/20210701010009/https://www.house.gov/representatives'
url_2022 = 'https://web.archive.org/web/20220701092720/https://www.house.gov/representatives'
sen_url_2017 = 'https://web.archive.org/web/20170430154244/https://www.senate.gov/general/committee_assignments/assignments.htm'
sen_url_2018 = 'https://web.archive.org/web/20170430154244/https://www.senate.gov/general/committee_assignments/assignments.htm'
sen_url_2019 = 'https://web.archive.org/web/20200223080341/https://www.senate.gov/general/committee_assignments/assignments.htm'
sen_url_2020 = 'https://web.archive.org/web/20200223080341/https://www.senate.gov/general/committee_assignments/assignments.htm'
sen_url_2021 = 'https://web.archive.org/web/20210311101643/https://www.senate.gov/general/committee_assignments/assignments.htm'
sen_url_2022 = 'https://web.archive.org/web/20210311101643/https://www.senate.gov/general/committee_assignments/assignments.htm'
```

```python
# make a request to the website and store the HTML content in a variable
response_2020 = requests.get(url_2020)
html_content_2020 = response_2020.content
```

```python
response_2021 = requests.get(url_2021)
html_content_2021 = response_2021.content
```

```python
response_2022 = requests.get(url_2022)
html_content_2022 = response_2022.content
```

```python
response_sen_2017 = requests.get(sen_url_2017)
html_content_sen_2017 = response_sen_2017.content
```

```python
response_sen_2018 = requests.get(sen_url_2018)
html_content_sen_2018 = response_sen_2018.content
```

```python
response_sen_2019 = requests.get(sen_url_2019)
html_content_sen_2019 = response_sen_2019.content
```

```python
response_sen_2020 = requests.get(sen_url_2020)
html_content_sen_2020 = response_sen_2020.content
```

```python
response_sen_2021 = requests.get(sen_url_2021)
html_content_sen_2021 = response_sen_2021.content
```

```python
response_sen_2022 = requests.get(sen_url_2022)
html_content_sen_2022 = response_sen_2022.content
```

```python
# parse the HTML content using BeautifulSoup
soup_2020 = BeautifulSoup(html_content_2020, 'html.parser')
soup_2021 = BeautifulSoup(html_content_2021, 'html.parser')
soup_2022 = BeautifulSoup(html_content_2022, 'html.parser')

soup_sen_2017 = BeautifulSoup(html_content_sen_2017, 'html.parser')
soup_sen_2018 = BeautifulSoup(html_content_sen_2018, 'html.parser')
soup_sen_2019 = BeautifulSoup(html_content_sen_2019, 'html.parser')
soup_sen_2020 = BeautifulSoup(html_content_sen_2020, 'html.parser')
soup_sen_2021 = BeautifulSoup(html_content_sen_2021, 'html.parser')
soup_sen_2022 = BeautifulSoup(html_content_sen_2022, 'html.parser')
```

```python
# fix anomily in 2022 data which has "|" as divider. REMOVE IT
soup_2022_str = soup_2022.prettify().replace("|", "")
soup_2022 = BeautifulSoup(soup_2022_str, 'html.parser')
```

```python
# find all tables on the soup object
tables_2020 = soup_2020.find_all('table')
tables_2021 = soup_2021.find_all('table')
tables_2022 = soup_2022.find_all('table')

tables_sen_2017 = soup_sen_2017.find_all('table')
tables_sen_2018 = soup_sen_2018.find_all('table')
tables_sen_2019 = soup_sen_2019.find_all('table')
tables_sen_2020 = soup_sen_2020.find_all('table')
tables_sen_2021 = soup_sen_2021.find_all('table')
tables_sen_2022 = soup_sen_2022.find_all('table')
```

```python
# create an empty list to store the dataframes
dataframes_2020 = []
dataframes_2021 = []
dataframes_2022 = []

dataframes_sen_2017 = []
dataframes_sen_2018 = []
dataframes_sen_2019 = []
dataframes_sen_2020 = []
dataframes_sen_2021 = []
dataframes_sen_2022 = []

```

```python
# loop through all tables and convert them to dataframes
for table in tables_2020:
    df = pd.read_html(str(table))[0]
    dataframes_2020.append(df)

for table in tables_2021:
    df = pd.read_html(str(table))[0]
    dataframes_2021.append(df)
    
for table in tables_2022:
    df = pd.read_html(str(table))[0]
    dataframes_2022.append(df)
    
for table in tables_sen_2017:
    df = pd.read_html(str(table))[0]
    dataframes_sen_2017.append(df)
    
for table in tables_sen_2018:
    df = pd.read_html(str(table))[0]
    dataframes_sen_2018.append(df)
    
for table in tables_sen_2019:
    df = pd.read_html(str(table))[0]
    dataframes_sen_2019.append(df)
    
for table in tables_sen_2020:
    df = pd.read_html(str(table))[0]
    dataframes_sen_2020.append(df)
    
for table in tables_sen_2021:
    df = pd.read_html(str(table))[0]
    dataframes_sen_2021.append(df)
    
for table in tables_sen_2022:
    df = pd.read_html(str(table))[0]
    dataframes_sen_2022.append(df)
```

```python
# concatenate the list of dataframes into one, along the rows (axis=0)
committee_2020 = pd.concat(dataframes_2020, axis=0, ignore_index=True)
committee_2021 = pd.concat(dataframes_2021, axis=0, ignore_index=True)
committee_2022 = pd.concat(dataframes_2022, axis=0, ignore_index=True)
```

```python
committee_sen_2017 = pd.concat(dataframes_sen_2017, axis=0, ignore_index=True)
```

```python

committee_sen_2017 = pd.concat(dataframes_sen_2017, axis=0, ignore_index=True)
committee_sen_2018 = pd.concat(dataframes_sen_2018, axis=0, ignore_index=True)
committee_sen_2019 = pd.concat(dataframes_sen_2019, axis=0, ignore_index=True)
committee_sen_2020 = pd.concat(dataframes_sen_2020, axis=0, ignore_index=True)
committee_sen_2021 = pd.concat(dataframes_sen_2021, axis=0, ignore_index=True)
committee_sen_2022 = pd.concat(dataframes_sen_2022, axis=0, ignore_index=True)
```
